```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tn.movie_budgets.csv')
```

```python
print(df.head())
```

```python
# Display the data types of the columns
print(df.dtypes)
```

```python
# Display unique values in the columns
print(df['production_budget'].unique()[:10])
print(df['domestic_gross'].unique()[:10])
print(df['worldwide_gross'].unique()[:10])
```

```python
# Remove non-numeric characters and convert to numeric
df['production_budget'] = df['production_budget'].replace('[\$,]', '', regex=True).astype(float)
df['domestic_gross'] = df['domestic_gross'].replace('[\$,]', '', regex=True).astype(float)
df['worldwide_gross'] = df['worldwide_gross'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Handle missing values if any
df = df.dropna(subset=['production_budget', 'domestic_gross', 'worldwide_gross'])
```

```python
# Create additional columns for profit and ROI
df['profit'] = df['worldwide_gross'] - df['production_budget']
df['roi'] = df['profit'] / df['production_budget']
```

```python
# Display the first few rows to see the new columns
print(df.head())
```

```python
# Data Analysis
# Calculate key statistics
avg_production_budget = df['production_budget'].mean()
avg_domestic_gross = df['domestic_gross'].mean()
avg_worldwide_gross = df['worldwide_gross'].mean()
avg_profit = df['profit'].mean()
avg_roi = df['roi'].mean()
```

```python
print(f'Average Production Budget: ${avg_production_budget:.2f}')
print(f'Average Domestic Gross: ${avg_domestic_gross:.2f}')
print(f'Average Worldwide Gross: ${avg_worldwide_gross:.2f}')
print(f'Average Profit: ${avg_profit:.2f}')
print(f'Average ROI: {avg_roi:.2f}')
```

```python
# Display the first few rows of the release_date column
print(df['release_date'].head())
```

```python
# Convert release date to datetime, handling errors
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
# Display the first few rows to ensure conversion
print(df['release_date'].head())
print(df.dtypes)
```

```python
# Drop rows with NaT in release_date after conversion
df = df.dropna(subset=['release_date'])
```

```python
# Extract the year from release_date
df['year'] = df['release_date'].dt.year
```

```python
# Display the first few rows to ensure the year column is created
print(df[['release_date', 'year']].head())
```

```python
print(df.head())
```

```python
#Group by year and calculate the mean only for numeric columns
annual_data = df.groupby('year').mean(numeric_only=True)[['production_budget', 'domestic_gross', 'worldwide_gross', 'profit', 'roi']]
```

```python
# Display annual data
print(annual_data.tail())
```

```python
# Bar chart of average gross earnings by release year
annual_data[['domestic_gross', 'worldwide_gross']].plot(kind='bar', figsize=(14, 6))
plt.title('Average Gross Earnings by Release Year')
plt.xlabel('Year')
plt.ylabel('Average Gross ($)')
plt.show()
```

```python
# Sort the DataFrame by ROI in descending order and select the top 10 movies
best_movies = df.sort_values(by='roi', ascending=False).head(10)
```

```python
# Visualize the top 10 movies based on ROI
plt.figure(figsize=(12, 6))
plt.bar(best_movies['movie'], best_movies['roi'], color='skyblue')
plt.xlabel('Movie Title')
plt.ylabel('Return on Investment (ROI)')
plt.title('Top 10 Movies based on ROI')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
