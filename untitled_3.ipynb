```python
import pandas as pd
```

```python
# Load the CSV file
file_path = 'bom.movie_gross.csv'
movies_df = pd.read_csv(file_path)
```

```python
print(movies_df.head())
```

```python
# Inspect the data types and check for missing values
print(movies_df.info())
print(movies_df.isnull().sum())
```

```python
import numpy as np
```

```python
# Inspect the columns
print(movies_df.columns)
```

```python
# Clean the column names
movies_df.columns = movies_df.columns.str.strip()
print(movies_df.columns)
```

```python
# Ensure the correct column names are used
domestic_gross_col = 'domestic_gross'
foreign_gross_col = 'foreign_gross'
```

```python
# Convert columns to numeric, coercing errors to NaN
movies_df[domestic_gross_col] = pd.to_numeric(movies_df[domestic_gross_col], errors='coerce')
movies_df[foreign_gross_col] = pd.to_numeric(movies_df[foreign_gross_col], errors='coerce')
```

```python
# Handle NaN values (option 1: fill NaN with 0)
movies_df[domestic_gross_col].fillna(0, inplace=True)
movies_df[foreign_gross_col].fillna(0, inplace=True)
```

```python
# Calculate total gross for each movie
movies_df['total_gross'] = movies_df[domestic_gross_col] + movies_df[foreign_gross_col]
```

```python
# Display the first few rows to verify
print(movies_df.head())
```

```python
# Group by 'title' to see performance of each title
title_gross = movies_df.groupby('title')['total_gross'].agg(['mean', 'median', 'sum']).reset_index()
```

```python
# Sort the title_gross dataframe by 'sum' (total gross) in descending order
title_gross_sorted = title_gross.sort_values(by='sum', ascending=False)

# Print out the top 10 titles
top_10_titles = title_gross_sorted.head(10)
print(top_10_titles)
```

```python
import matplotlib.pyplot as plt
```

```python
# Create a bar chart of the top 10 titles by total gross
plt.figure(figsize=(12, 6))
plt.bar(top_10_titles['title'], top_10_titles['sum'], color='skyblue')
plt.xlabel('Title')
plt.ylabel('Total Gross')
plt.title('Top 10 Titles by Total Gross')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
